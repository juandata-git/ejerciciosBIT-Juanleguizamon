# Actividad 5: Taller SQL (SQL Server)

Para completar el taller recuerda estas sentencias de SQL:

- DML (modifica datos):

  - INSERT: agrega filas a una tabla.
  - UPDATE: modifica columnas de filas existentes.
  - DELETE: elimina filas.
- SELECT (consulta datos):
  - SELECT: elige columnas a devolver (puede incluir expresiones).
  - WHERE: filtra filas (antes de agrupar).
  - GROUP BY: agrupa filas para usar agregados (SUM, COUNT, AVG, MIN, MAX).
  - HAVING: filtra grupos (después de GROUP BY).
  - ORDER BY: ordena el resultado.
  - TOP (N): limita la cantidad de filas devueltas (útil con ORDER BY).

### 4. Lista ProductID, p.Name AS ProductName, c.Name AS CategoryName para productos con categoría
(INNER JOIN).

```python
respuesta4 = pd.read_sql_query(
    """
    SELECT ProductID, p.Name AS ProductName, c.Name AS CategoryName
    FROM SalesLT.Product AS p
    INNER JOIN SalesLT.ProductCategory AS c ON p.ProductCategoryID = c.ProductCategoryID
    ORDER BY ProductID
    """,
    conn
)
respuesta4
```

### 5. Obtén pedidos del año 2008 en adelante (o el año que aplique en tu base) con SalesOrderID, OrderDate, CustomerID, TotalDue y el nombre del cliente (FirstName + ' ' + LastName).
(JOIN + WHERE)

```python
respuesta5 = pd.read_sql_query(
    """
    SELECT
        SalesLT.SalesOrderHeader.SalesOrderID,
        SalesLT.SalesOrderHeader.OrderDate,
        SalesLT.SalesOrderHeader.CustomerID,
        SalesLT.SalesOrderHeader.TotalDue,
        SalesLT.Customer.FirstName + ' ' + SalesLT.Customer.LastName AS CustomerName
    FROM SalesLT.SalesOrderHeader
    INNER JOIN SalesLT.Customer ON SalesLT.SalesOrderHeader.CustomerID = SalesLT.Customer.CustomerID
    WHERE YEAR(SalesLT.SalesOrderHeader.OrderDate) >= 2008
    ORDER BY SalesLT.SalesOrderHeader.OrderDate
    """,
    conn
)
respuesta5
```

### 6. Para cada ProductID, suma la cantidad vendida (SUM(OrderQty)). Devuelve solo productos con ventas totales > 1000, ordenados de manera descendente.
Usa GROUP BY, HAVING y ORDER BY

```python
respuesta6 = pd.read_sql_query(
    """
    SELECT
        ProductID,
        SUM(OrderQty) AS TotalVendido
    FROM SalesLT.SalesOrderDetail
    GROUP BY ProductID
    HAVING SUM(OrderQty) > 1000
    ORDER BY TotalVendido DESC
    """,
    conn
)
respuesta6


## No hay productos con mas de 1000 unidades vendidas (busque y la maxima cantidad es 87)
```

### 7. Productos sin ventas: lista productos que no aparecen en SalesLT.SalesOrderDetail.
Usa LEFT JOIN + WHERE

```python
respuesta7 = pd.read_sql_query(
    """
    SELECT
        SalesLT.Product.ProductID,
        SalesLT.Product.Name
    FROM SalesLT.Product
    LEFT JOIN SalesLT.SalesOrderDetail
        ON SalesLT.Product.ProductID = SalesLT.SalesOrderDetail.ProductID
    WHERE SalesLT.SalesOrderDetail.ProductID IS NULL
    """,
    conn
)
respuesta7
```

### 8. Clientes por país: total de clientes por CountryRegion (usa Address + CustomerAddress).
Podrias necesitar: COUNT(DISTINCT CustomerID).

```python
respuesta8 = pd.read_sql_query(
    """
    SELECT
        SalesLT.Address.CountryRegion,
        COUNT(DISTINCT SalesLT.CustomerAddress.CustomerID) AS TotalClientes
    FROM SalesLT.CustomerAddress
    INNER JOIN SalesLT.Address
        ON SalesLT.CustomerAddress.AddressID = SalesLT.Address.AddressID
    GROUP BY SalesLT.Address.CountryRegion
    ORDER BY TotalClientes DESC
    """,
    conn
)
respuesta8
```

### 9. Ticket promedio por año: AVG(TotalDue) por YEAR(OrderDate).

```python
respuesta9 = pd.read_sql_query(
    """
    SELECT
        YEAR(OrderDate) AS Año,
        AVG(TotalDue) AS TicketPromedio
    FROM SalesLT.SalesOrderHeader
    GROUP BY YEAR(OrderDate)
    ORDER BY Año
    """,
    conn
)
respuesta9
```
