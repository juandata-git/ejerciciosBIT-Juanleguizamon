```python


import pandas as pd

# 1Ô∏è‚É£ Leer el archivo CSV
ruta = r"C:\Users\juanl\Downloads\ballondor_2025_nominees_dataset.csv"
df = pd.read_csv(ruta)

# 2Ô∏è‚É£ Mostrar forma, info general y estad√≠sticas
print("üîπ Forma del dataset:", df.shape)
print("\nüîπ Informaci√≥n general:")
print(df.info())
print("\nüîπ Estad√≠sticas descriptivas:")
print(df.describe(include='all'))

# 3Ô∏è‚É£ Generar nuevo dataset con filtro (ejemplo: jugadores con m√°s de 20 goles)
df_filtrado = df[df['Gls'] > 20]
print("\nüîπ Dataset filtrado (Goles > 20):")
print(df_filtrado[['Player', 'Squad', 'Gls']])

# 4Ô∏è‚É£ Generar nuevo dataset con ordenamiento (ejemplo: por asistencias descendente)
df_ordenado = df.sort_values(by='Ast', ascending=False)
print("\nüîπ Dataset ordenado por asistencias:")
print(df_ordenado[['Player', 'Squad', 'Ast']])

df_filtrado.to_csv("filtrado_goles.csv", index=False)
df_ordenado.to_csv("ordenado_asistencias.csv", index=False)

df_filtrado.head()
df_ordenado.head()
```
