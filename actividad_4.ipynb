```python
import pandas as pd
import numpy as np

# lectura de archivo

pizza = pd.read_excel("Enhanced_pizza_sell_data_2024-25.xlsx")

pizza.head()



```

```python
##renombrando las columnas de ingles a español

pizza.rename(columns={
    "Order ID": "ID Pedido",
    "Restaurant Name": "Restaurante",
    "Location": "Ubicación",
    "Order Time": "Hora Pedido",
    "Delivery Time": "Hora Entrega",
    "Delivery Duration (min)": "Duración Entrega (min)",
    "Pizza Size": "Tamaño Pizza",
    "Pizza Type": "Tipo Pizza",
    "Toppings Count": "Cantidad Toppings",
    "Distance (km)": "Distancia (km)",
    "Traffic Level": "Nivel Tráfico",
    "Payment Method": "Método Pago",
    "Is Peak Hour": "Es Hora Pico",
    "Is Weekend": "Es Fin de Semana",
    "Delivery Efficiency (min/km)": "Eficiencia Entrega (min/km)",
    "Topping Density": "Densidad Toppings",
    "Order Month": "Mes Pedido",
    "Payment Category": "Categoría Pago",
    "Estimated Duration (min)": "Duración Estimada (min)",
    "Delay (min)": "Retraso (min)",
    "Is Delayed": "Está Retrasado",
    "Pizza Complexity": "Complejidad Pizza",
    "Traffic Impact": "Impacto Tráfico",
    "Order Hour": "Hora del Pedido",
    "Restaurant Avg Time": "Tiempo Promedio Restaurante"
}, inplace=True)

pizza
```

```python
# Eliminar filas con datos faltantes en columnas clave
# → Asegura que no haya valores nulos en columnas necesarias para el análisis
pizza.dropna(subset=["Nivel Tráfico", "Complejidad Pizza", "Distancia (km)", "Retraso (min)"], inplace=True)

# Reemplazar valores negativos de retraso por 0
# → Corrige errores: un retraso no puede ser negativo
pizza["Retraso (min)"] = pizza["Retraso (min)"].apply(lambda x: max(x, 0))

# Filtrar pedidos con distancia válida
# → Elimina registros con distancia 0 o negativa, que no tienen sentido para el análisis
pizza = pizza[pizza["Distancia (km)"] > 0]

# Seleccionar columnas relevantes
# → Crea un subconjunto del DataFrame con solo las variables que vamos a analizar
factores = pizza[["Nivel Tráfico", "Complejidad Pizza", "Es Hora Pico", "Es Fin de Semana",
                  "Distancia (km)", "Cantidad Toppings", "Retraso (min)"]]

# Crear variable: retraso por kilómetro
# → Mide la eficiencia del tiempo de entrega en función de la distancia
pizza["Retraso por km"] = pizza["Retraso (min)"] / pizza["Distancia (km)"]

# Crear variable binaria: ¿retraso alto?
# → Marca los pedidos con más de 10 minutos de retraso como "altos"
pizza["Retraso Alto"] = pizza["Retraso (min)"] > 10

# Retraso promedio por categoría
# → Calcula el promedio de retraso según cada factor categórico
retraso_trafico = pizza.groupby("Nivel Tráfico")["Retraso (min)"].mean()
retraso_complejidad = pizza.groupby("Complejidad Pizza")["Retraso (min)"].mean()
retraso_pico = pizza.groupby("Es Hora Pico")["Retraso (min)"].mean()
retraso_finde = pizza.groupby("Es Fin de Semana")["Retraso (min)"].mean()

import numpy as np

# Normalizar retraso
# → Convierte los valores de retraso a escala estándar (z-score) para comparar mejor
retrasos = pizza["Retraso (min)"].values
media = np.mean(retrasos)
desvio = np.std(retrasos)
pizza["Retraso Normalizado"] = (retrasos - media) / desvio

# Correlaciones con variables numéricas
# → Mide qué tan relacionadas están las variables numéricas con el retraso
correlaciones = pizza[["Retraso (min)", "Distancia (km)", "Cantidad Toppings"]].corr()

# Mostrar el DataFrame final
pizza
```
